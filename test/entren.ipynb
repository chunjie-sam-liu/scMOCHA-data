```python
from Bio import Entrez

# Set your email (required by NCBI)
Entrez.email = "chunjie.sam.liu@gmail.com"

# Fetch BioProject metadata
handle = Entrez.esearch(db="bioproject", term="PRJNA616380")
record = Entrez.read(handle)
project_id = record["IdList"][0]

# Fetch detailed BioProject record
handle = Entrez.esummary(db="bioproject", id=project_id)
project_details = Entrez.read(handle)
print(project_details)
```

```python
import requests

# Fetch SRA Run IDs from NCBI
url = "https://www.ncbi.nlm.nih.gov/sra/?term=PRJNA616380"
response = requests.get(f"{url}&report=runinfo&format=json")
data = response.json()
```

```python

# Extract Run IDs
run_ids = [item["Run"] for item in data]
print(run_ids)
```
